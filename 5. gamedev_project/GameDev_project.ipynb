# Сборный проект 

**Описание проекта**

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 

**Целью исследования** является выявить определяющие успешность игры закономерности, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

В качестве **материала исследования** используются данные до 2016 года. В процессе исследования необходимо отработать принцип работы с данными и оценить положение дел в игровой индустрии.

**Описание данных**

- `Name` — название игры
- `Platform` — платформа
- `Year_of_Release` — год выпуска
- `Genre` — жанр игры
- `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
- `EU_sales` — продажи в Европе (миллионы проданных копий)
- `JP_sales` — продажи в Японии (миллионы проданных копий)
- `Other_sales` — продажи в других странах (миллионы проданных копий)
- `Critic_Score` — оценка критиков (максимум 100)
- `User_Score` — оценка пользователей (максимум 10)
- `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

- в столбцах `Critic_Score`, `User_score` и `Rating` отсутсвуют почти 50% данных;
- в наименовании столбцов допущена стилевая ошибка - смешение регистров.

**Предположения касательно пропусков:**

- в столбце `Critic_Score` числовые данные с оценкой критиков, но есть вероятность, что оценки отсутствуют у более старых игр, т.к. в то время либо не было инструментов для определения рейтинга, либо на тот момент не было у критиков такой привычки - игры оценивать;
- пропуски в столбце `Rating` могут быть обусловлены тем, что до какого-то момента организации Entertainment Software Rating Board не существовало, хотя она основана в 1994 г. Неужели процент старых игр в выборке настолько высок?
- Со столбцом `User_Score` ситуация аналогичная `Critic_Score`;
- пропуски в столбце `Year_of_Release` обучсловлены либо человеческим фактором, либо ошибкой при выгрузке, т.к. процент таких пропусков достаточно мал, плюс, эту информацию проще достоверно установить элементарным поиском, а вот с рейтингом так уже не сработает.

**Что можно предпринять:**

- данные в столбце `Year_of_Release` можно оставить без каких бы то ни было преобразований, т.к., во-первых, пропусков там чуть больше 1%, во-вторых, пропуски не помешают расчетам, т.к. при расчете среднего и т.д. учитываются числовые данные;
- пропуски в столбцах с рейтингом от критиков и пользователей тоже можно оставить в виде NaN, т.к. если мы присвоим пропускам значение 0, например, то смешаем им с теми играми, у которых действительный рейтинг равен нулю. Среднее посчитать тоже нельзя, даже если возьмем рейтинги в отдельном конкретном жанре, т.к. слишком много факторов влияет на оценку той или иной игры;
- в столбце `Rating` также не можем воспользоваться медианой для заполнения пропусков по причине указанной в прошлои пункте. Лучше использоват значение `No ESRB` или `Rating is not defined`.

## Предобработка данных

### Приведем названия всех столбцов к нижнему регистру

```python
games.columns = games.columns.str.lower()
```

### Изменим тип данных в `user_score`

Посмотрим, чем может быть обусловлен такой вид данных, возможно, в столбце есть какие-то значения помимо числовых.

```python
games['user_score'].unique()
```

Помимо NaN'ов видим значение 'tbd'.

Если верить Кэмбридскому словарю, ["tbd - аббревиатура, расшифровывающаяся как to be decided or to be determined"](https://dictionary.cambridge.org/ru/%D1%81%D0%BB%D0%BE%D0%B2%D0%B0%D1%80%D1%8C/%D0%B0%D0%BD%D0%B3%D0%BB%D0%B8%D0%B9%D1%81%D0%BA%D0%B8%D0%B9/tbd). Т.е. иными словами это тот же NaN, ведь для таких игр еще не установлен рейтинг по той или иной причине (например, недостаточное кол-во оценок пользователей).

```python
# заменить значения 'tbd' на Nan
games['user_score'] = games['user_score'].replace('tbd', np.NaN)
games['user_score'] = games['user_score'].astype('float')
```

### Проверка на дубликаты

```python
# проверим, есть ли явные дубликаты
games.duplicated().sum()
```

Явных дубликатов нет.

### Замена пропусков в столбце `rating`

```python
games['rating'] = games['rating'].fillna('No ESRB')
```

### Добавление столбца с суммой продаж по всем регионам

```python
# создадим новый столбец таблицы, куда сохраним сумму вырученных с продажи средств по регионам
#games['sales_total'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
games['sales_total'] = games.filter(like='sales').sum(axis=1)
games.head()
```

## Исследовательский анализ данных

### Кол-во выпущенных игр по годам

```python
# построим график с распределением кол-ва выпущенных игр по годам
# сначала составим сводную таблицу с кол-ом игр, выпущенных в определенный год
num_of_games_per_year = games.pivot_table(index='year_of_release', values='name', aggfunc='count')
num_of_games_per_year.columns = ['count']
# на основе сводной построим таблицу
num_of_games_per_year.plot(kind='bar', figsize=(15,7), ec='black')
plt.xlabel('Год выпуска')
plt.ylabel('Кол-во выпущенных игр')
plt.title('Распределение выпуска игр по годам')
```

### Суммарные продажи игр по платформам

```python
# создадим сводную таблицу с суммарной прибылью от игр, выпущенных на платформе
total_sales_by_platform = games.pivot_table(index='platform', values='sales_total', aggfunc='sum')
total_sales_by_platform = total_sales_by_platform.sort_values(by='sales_total', ascending=True)
# построим график
total_sales_by_platform.plot(kind='barh', figsize=(15,15), ec='black', legend=True)
plt.xlabel('Суммарная прибыль')
plt.ylabel('Платформа')
plt.title('Продано игр в млн. копий')
```

### Динамика продаж игр по платформам

Посмотрим, как менялись продажи по платформам. Выберем платформы с наибольшими суммарными продажами и построим распределение по годам. Ответим на вопрос: за какой характерный срок появляются новые и исчезают старые платформы?

Как мы видим из пункта 3.2. в топ 6 платформ с наибольшими суммарными продажами входят PS2, X360, PS3, Wii, DS, PS. 

#### Распределение прибыли по годам для PS2

```python
# соберем в список наименования платформ, которые лидируют по кол-ву проданных копий игр
top_platforms = ['PS2', 'X360', 'PS3', 'Wii', 'DS', 'PS']
# построим общий график для всех шести платформ, с динамикой кол-ва продаж в зависимости от года
fig, ax = plt.subplots(figsize=(15, 10))
for pl in top_platforms:
    (games[games['platform'] == pl]
     .groupby('year_of_release')['sales_total'].sum()
     .plot(grid=True, label=pl, ax=ax, style='o-', legend=True)
    )
plt.title('Динамика продаж по годам среди платформ - лидеров по продажам')
plt.xlabel('Год')
plt.ylabel('Продажи, млн. проданных копий')
plt.show()
```

PS2 продержалась на рынке 11 лет - с 2000 по 2011 гг.

Платформа X360 также продержалась на рынке 11 лет - с 2005 по 2016 гг.

PS3 с 2006 по настоящее время ещё живёт, но начала сильно сдавать позиции с выходом в 2013 году PS4.

Wii на рынке с 2006 по нынешний день, но к 2016, спустя 10 лет, уже почти изжила себя.

DS вышла в 1985, но за период с 1986 по 2003 гг. нет данных, скорее всего, по той причине, что они не собирались. И с 2004 года популярность стала расти, пик в 2008, конец срока службы - 2013 год - 8 активных лет жизни.

PS прожила с 1994 года до 2003, т.е. почти 10 лет.

Таким образом, делаем вывод о том, что в среднем платформа живет 10 лет.

```python
def create_sales_boxplot(title=None, query_condition = None, ylims=None):
    %matplotlib inline
    if query_condition == None:
        ax = games.boxplot(column='sales_total', fontsize=25, figsize=(25,10),
                                         boxprops=dict(linestyle='-', linewidth=5),
                                         medianprops=dict(linestyle='-', linewidth=5),
                                         whiskerprops=dict(linestyle='-', linewidth=5),
                                         capprops=dict(linestyle='-', linewidth=5), 
                                         notch=True, showfliers=False)
    else:
        df_query = games.query(query_condition)
        by_column = query_condition.split()[0]
        ax = df_query.boxplot(column='sales_total', by=by_column, fontsize=25, figsize=(25,10),
                                         boxprops=dict(linestyle='-', linewidth=5),
                                         medianprops=dict(linestyle='-', linewidth=5),
                                         whiskerprops=dict(linestyle='-', linewidth=5),
                                         capprops=dict(linestyle='-', linewidth=5),
                                         notch=True, showfliers=False)
        ax.set_xlabel('')
        ax.set_title('')
    if ylims != None:
        ax.set_ylim(ylims)
    if title != None:
        plt.suptitle(title, y=0.94, fontsize=25, fontweight='bold')
```

## Проверка Гипотез

Перед нами стоит задача проверить 2 гипотезы.

**Гипотезы:**

- средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- средние пользовательские рейтинги жанров Action и Sports разные.

Для проверки обеих гипотез будем использовать t-критерий Стьюдента, т.к. в обоих случаях выборки независимы между собой.
Определим пороговое значение alpha = 0.05

### Гипотеза № 1

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

- Н0 (нулевая гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
- Н1 (альтернативная гипотеза) - средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

```python
xbox_one_ratings = games[(games['platform'] == 'XOne') & (games['user_score'] > 0)]['user_score']
pc_ratings = games[(games['platform'] == 'PC') & (games['user_score'] > 0)]['user_score']
alpha = 0.05
stat, p = ttest_ind(xbox_one_ratings, pc_ratings)

print('t = {}\np-value = {:.10e}'.format(stat, p))
print('\nПороговое значение alpha = 0.05')

if p < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Отвергает нулевую гипотезу, а это значит, что средние пользовательские рейтинги платформ Xbox One и PC отличаются.

Посмотрим, в чью пользу отличие.

```python
print('Средняя оценка пользователей игр для платформы Xbox One  - {:.2f}'.format(xbox_one_ratings.mean()))
print('Средняя оценка пользователей игр для платформы PC - {:.2f}'.format(pc_ratings.mean()))
```

Игры на PC получают оценки статистически значимо выше, чем игры для Xbox One. 

### Гипотеза № 2

Средние пользовательские рейтинги жанров Action и Sports разные.

- Н0 (нулевая гипотеза) - средние пользовательские рейтинги жанров Action и Sports одинаковые.
- Н1 (альтернативная гипотеза) - средние пользовательские рейтинги жанров Action и Sports отличаются между собой.

```python
action_ratings = games[(games['genre'] == 'Action') & (games['user_score'] > 0)]['user_score']
pc_ratings = games[(games['genre'] == 'Sports') & (games['user_score'] > 0)]['user_score']
stat, p = ttest_ind(action_ratings, pc_ratings)
alpha = 0.05

print('t = {}\np-value = {:.4f}'.format(stat, p))
print('\nПороговое значение alpha = 0.05')

if p < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Не можем отвергнуть нулевую гипотезу, а значит, средние пользовательские рейтинги жанров Action и Sports одинаковые
